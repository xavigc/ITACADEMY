```python
def convert_temp(temp, unitats_mesura):
  
    if unitats_mesura == 'C':
        if temp > 60 or temp < (-25):
            print ('temperatura Celsius fora de rang')
            return
        else:
            temp_celsius = temp
            temp_fahrenheit = round((temp * 9 / 5) + 32,2)
            temp_kelvin     = round((temp + 273.15),2)
    elif unitats_mesura == 'F':
        if temp > 140 or temp < (-13):
            print ('temperatura Fahrenheit fora de rang')
            return
        else:
            temp_fahrenheit = temp 
            temp_celsius    = round((temp - 32) * 5/9,2)
            temp_kelvin     = round((temp_celsius + 273.15),2)
    elif unitats_mesura == 'K':
        if temp > 333.15 or temp < (-248):
            print ('temperatura Kelvin fora de rang')
            return
        else:
            temp_kelvin     = temp
            temp_celsius    = round((temp - 273.15),2)
            temp_fahrenheit = round((temp_celsius * 9 / 5) + 32,2)
    else:
        print("Unitat de mesura no es Celsius, Fahrenheit o Kelvin (C, F o K) ")     
        return 0,0 # Retornem 0 com a resultat de la funció
    return 1,temp_celsius,temp_fahrenheit, temp_kelvin

while True:
    temp = input("Dona'm una temperatura")
    try:
        temp = int(temp)
        break
    except:
        print ("Error valor no numéric")
while True:
    unitats_mesura = input("Indica l'unitat d'aquesta temperatura Celsius, Fahrenheit o Kelvin (C, F o K)" )
    try:
        unitats_mesura = unitats_mesura.upper()
        break
    except:
        print ("Error valor no nstring")
    
temperaturas = convert_temp(temp,unitats_mesura)

if temperaturas[0] == 1: #Controlem l'estas que retorna la funció
    if unitats_mesura == 'C':
        print("Temperatura donada", temp, "graus Celsius", "equival a", temperaturas[2],"Fahrenheit i",temperaturas[3],"Kelvin")
    elif unitats_mesura == 'F':
        print("Temperatura donada", temp, "graus Fahrenheit", "equival a",temperaturas[1],"Celsius i",temperaturas[3],"Kelvin")
    elif unitats_mesura == 'K':
        print("Temperatura donada", temp, "graus Kelvin", "equival a",temperaturas[1],"Celsius i",temperaturas[2],"Fahrenheit")


```

```python
# funcions de conversió de temperatura

def c_a_f(c):
    return c * 9/5 + 32

def c_a_k(c):
    return c + 273.15

def f_a_c(f):
    return (f - 32) * 5/9

def f_a_k(f):
    return (f - 32) * 5/9 + 273.15

def k_a_c(k):
    return k - 273.15

def k_a_f(k):
    return (k - 273.15) * 9/5 + 32

conversions =  {("C", "F"): c_a_f,("C", "K"): c_a_k,("F", "C"): f_a_c,("F", "K"): f_a_k,("K", "C"): k_a_c,("K", "F"): k_a_f}

unitats = {"C", "F", "K"}


def validar_unitat(unitat):
    if unitat not in unitats:
        raise ValueError("Unitat no vàlida. Utilitza: C, F, K")


def validar_temperatura(valor):
    try:
        return float(valor)
    except ValueError:
        raise ValueError("La temperatura ha de ser un numero")


def convertir(temperatura, origen):
    #Retorna un diccionari amb totes les conversions possibles excepte la unitat d'origen
    resultats = {}

    for (u_origen, u_desti), funcio in conversions.items():
        if u_origen == origen:
            resultats[u_desti] = funcio(temperatura)

    return resultats


def main():
    try:
        temp_input = input("Introdueix la temperatura: ")
        temperatura = validar_temperatura(temp_input)

        origen = input("Introdueix la unitat d'origen (C, F, K): ").upper()
        validar_unitat(origen)

        conversions = convertir(temperatura, origen)

        print("Conversions de", temperatura, origen,":")
        for unitat, valor in conversions.items():
            print(f"{valor:.2f} {unitat}")

    except ValueError as error:
        print(f"Error: {error}")


if __name__ == "__main__":
    main()
```

```python
def repeticions_paraula(text):
    paraules = text.split()
    comptador = {}
    for paraula in paraules:
        paraula = paraula.lower()
        if paraula in comptador:
            comptador[paraula] += 1
        else:
            comptador[paraula] = 1
    return comptador

def validar_text(text):
    import string
    # Comprovar que l'entrada és un text
    if not isinstance(text, str):
        raise ValueError("L'entrada ha de ser un text (string)")

    # Comprovar que el text no està buit
    if not text.strip():
        raise ValueError("El text no pot estar buit")
    
    # Eliminar signes de puntuació
    for signe in string.punctuation:
        text = text.replace(signe, "")
    return text

def llargada_mitjana(text):
    paraules = text.split()
    total_lletres = sum(len(paraula) for paraula in paraules)
    llargada_mitjana = total_lletres / len(paraules)
    print(f"Llargada mitjana de les paraules: {llargada_mitjana:.0f} lletres")

def main():
    try:
        text = '''Lorem ipsum dolor sit amet consectetur adipiscing elit, dui platea pulvinar euismod rutrum inceptos commodo lacus, dictum nostra montes at congue nam. 
                  Faucibus tincidunt ultricies parturient sociis fusce quam leo, ultrices per ligula penatibus mollis sapien nisi dui, nullam rutrum elementum blandit aliquam imperdiet. 
                  Nibh tortor litora magnis feugiat lectus parturient urna etiam gravida, mauris magna sollicitudin justo tellus netus curae.'''    
        text = validar_text(text)
        resultats = repeticions_paraula(text)       
        print("Repeticions de paraules:")
        for paraula, compta in resultats.items():
            print(f"'{paraula}': {compta} vegades") 
        llargada_mitjana(text)

    except ValueError as error:
        print(f"Error: {error}")

if __name__ == "__main__":
    main()
```

```python
def invertir_diccionari(diccionari):
    invertit = {}
    for clau, valor in diccionari.items():
        if valor in invertit:
            invertit[valor].append(clau)
        else:
            invertit[valor] = [clau]
    duplicats = {valor: clau for valor, clau in invertit.items() if len(clau) > 1}

    if duplicats:
        print("Elements duplicats en el diccionari invertit:", invertit)
    else:
        print("Diccionari invertit:", invertit)

def main():
    try:
        diccionari = {"a": 1, "b": 2, "c": 3}
        invertit = invertir_diccionari(diccionari)
        diccionari =  {'x':'apple', 'y':'banana', 'z':'banana'} 
        invertit = invertir_diccionari(diccionari)
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```
